```python
import os
import json
import hashlib
from itertools import chain

DATA_LOC = '/root/xlcoder/MiniMind2-Small/dataset'
input_path = os.path.join(DATA_LOC, "sft_data_en.jsonl")
output_path = os.path.join(DATA_LOC,"cls_sft.jsonl")

def keep_line(text, mod=9):
    h = hashlib.md5(text.encode("utf-8")).hexdigest()
    return int(h, 16) % mod == 0

with open(input_path, "r", encoding="utf-8") as fin, \
     open(output_path, "w", encoding="utf-8") as fout:
    for line in fin:
        if keep_line(line, mod=10):
            try:
                obj = json.loads(line)
                fout.write(json.dumps(
                    {"conversations": 
                     [{"role":"user", 
                       "content": ' '.join(list(chain.from_iterable(obj['history']))+[obj['input']])}, 
                      {"role": "assistant", "content": obj['output']}]},
                    ensure_ascii=False
                ) + "\n")
            except json.JSONDecodeError:
                continue
```
