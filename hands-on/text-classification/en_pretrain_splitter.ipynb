```python
import os
import json
import hashlib
from itertools import chain

DATA_LOC = '/root/xlcoder/MiniMind2-Small/dataset'
input_path = os.path.join(DATA_LOC, "sft_data_en.jsonl")
output_path = os.path.join(DATA_LOC,"subset_output_10pct.jsonl")

def keep_line(text, mod=10):
    h = hashlib.md5(text.encode("utf-8")).hexdigest()
    return int(h, 16) % mod == 0

with open(input_path, "r", encoding="utf-8") as fin, \
     open(output_path, "w", encoding="utf-8") as fout:
    for line in fin:
        if keep_line(line, mod=10):
            try:
                obj = json.loads(line)
                res = ' '.join(list(chain.from_iterable(obj['history']))+[obj['input']]+[obj['output']])
                fout.write(json.dumps(
                    {"text": res},
                    ensure_ascii=False
                ) + "\n")
            except json.JSONDecodeError:
                continue
```
